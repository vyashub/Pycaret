# Model Ensembling
Ensembling approaches train several classifiers in the hope that combining their predictions will outperform any single classifier (worst case scenario, be better than the worse classifier). The combination rule can be: majority vote, mean, max, min, product … the average rule is the most used.

## Ensemble Model
https://pycaret.org/ensemble-model/

https://quantdare.com/what-is-the-difference-between-bagging-and-boosting/

```python
from pycaret.datasets import get_data
diabetes = get_data('diabetes')

# Importing module and initializing setup
from pycaret.classification import *
clf1 = setup(data = diabetes, target = 'Class variable',silent=True)

# creating decision tree model
dt = create_model('dt')
# ensembling decision tree model (boosting)
dt_bagged = ensemble_model(dt,method="Boosting")
```

# Blend Models

https://mlwave.com/kaggle-ensembling-guide/

```python
# Importing dataset
from pycaret.datasets import get_data
boston = get_data('boston')

# Importing module and initializing setup
from pycaret.regression import *
reg1 = setup(data = boston, target = 'medv',silent=True)

# creating multiple models for blending
dt = create_model('dt')
catboost = create_model('catboost')
lightgbm = create_model('lightgbm')

# blending specific models
blender = blend_models(estimator_list = [dt, catboost, lightgbm])
```

# Stack Model

```python
# Importing dataset
from pycaret.datasets import get_data
diabetes = get_data('diabetes')

# Importing module and initializing setup
from pycaret.classification import *
clf1 = setup(data = diabetes, target = 'Class variable',silent=True)

# create individual models for stacking
ridge = create_model('ridge')
lda = create_model('lda')
gbc = create_model('gbc')
xgboost = create_model('xgboost')

# stacking models
stacker = stack_models(estimator_list = [ridge,lda,gbc], meta_model = xgboost)
```
